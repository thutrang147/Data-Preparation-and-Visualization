```python
import seaborn as sns
import pandas as pd
titanic_data = sns.load_dataset('titanic')
```

```python
titanic_data.head()
```

```python
titanic_data_sorted = titanic_data.sort_values(by='age', ascending=False)
```

```python
titanic_data_sorted.head()
```

```python
survival_rate = titanic_data['survived'].mean()
print(survival_rate)
```

```python
def age_group(age):
    if pd.isna(age):
        return 'Unknown'
    elif age < 18:
        return 'Child'
    elif age < 65:
        return 'Adult'
    else:
        return 'Senior'

titanic_data_sorted['agegroup'] = titanic_data_sorted['age'].apply(age_group)
```

```python
titanic_data_sorted.head()
```

```python
columns = titanic_data_sorted.select_dtypes(include=['float64', 'int64'])
columns.corr()
```
