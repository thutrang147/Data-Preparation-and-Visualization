Lab #1: apply and map

```python
import pandas as pd
```

```python
data = { 'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
'Score': [88, 76, 95, 65, 70],
'Days_Off': [2, 5, 1, 8, 6]
}
df_lab1 = pd.DataFrame(data)
```

```python
df_lab1
```

```python
def evaluate_performance(row):
    if row['Score'] >= 85 and row['Days_Off'] <= 3:
        return 'Excellent'
    elif row['Score'] >= 70 and row['Days_Off'] <= 5:
        return 'Good'
    elif row['Score'] >= 50:
        return 'Average'
    else:
        return 'Poor'

df_lab1['Performance'] = df_lab1.apply(evaluate_performance, axis=1)
```

```python
df_lab1
```

```python
def categorize_score(score):
    return 'High' if score >= 80 else 'Low'

df_lab1['Score_Category'] = df_lab1['Score'].map(categorize_score)
df_lab1
```

```python
import seaborn as sns
titanic_data = sns.load_dataset('titanic')
```

```python
titanic_data.head()
```

```python
titanic_data.sort_values(by='age', ascending=False, inplace=True)
```

```python
titanic_data
```

```python
overall_survival_rate = titanic_data['survived'].mean()
overall_survival_rate
```

```python
def age_group(age):
    if age < 18:
        return 'Child'
    elif age < 65:
        return 'Adult'
    else:
        return 'Senior'
    
titanic_data['age_group'] = titanic_data['age'].map(age_group)
```

```python
columns = titanic_data.select_dtypes(include=['float64', 'int64'])
columns.corr()
```
