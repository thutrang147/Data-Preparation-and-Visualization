##### **Exercise 13: Testing the 80/20 Principle (Customers and Products)**

```python
# a 
customer_revenue = cleaned_retail.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False)
total_revenue = customer_revenue.sum()
cumulative_pct = (customer_revenue.cumsum() / total_revenue * 100)
customers_for_80 = len(cumulative_pct[cumulative_pct <= 80])
total_customers = len(customer_revenue)
pct_customers_for_80 = (customers_for_80 / total_customers) * 100
pct_customers_for_80
```

```python
# b 
products_revenue = cleaned_retail.groupby('StockCode')['TotalPrice'].sum().sort_values(ascending=False)
total_rev_products = products_revenue.sum()
cumulative_pct_products = (products_revenue.cumsum() / total_rev_products * 100)
products_80 = len(cumulative_pct_products[cumulative_pct_products <= 80])
total_products = len(products_revenue)
pct_products_80 = (products_80 / total_products) * 100
pct_products_80
```

```python
# c 
''' 
Based on the results above (customers: 26% and products: 21%) -> 80/20 principle 
does not hole true for both in this dataset
'''
```

##### **Exercise 15: Analysis - Behavior of VIP Customerss**

```python
# a
rev_by_customer = cleaned_retail.groupby('CustomerID')['TotalPrice'].sum()
rev_by_customer.quantile(0.95)
```

```python
vip_customers = cleaned_retail[cleaned_retail['TotalPrice'] > rev_by_customer.quantile(0.95)]['CustomerID']
vip_customers
```

```python
# b 
top_products = cleaned_retail[cleaned_retail['CustomerID'].isin(vip_customers)]['StockCode'].value_counts().sort_values(ascending=False).head(10)
top_products
```

```python
# c 
vip_aov = cleaned_retail[cleaned_retail['CustomerID'].isin(vip_customers)].groupby('InvoiceNo')['TotalPrice'].sum().mean()
vip_aov
```

```python
vip_aov > overall_aov
```

```python
# d 
top_vip_products = cleaned_retail[cleaned_retail['CustomerID'].isin(vip_customers)][['StockCode', 'Description']].drop_duplicates()
top_vip_products = top_vip_products[top_vip_products['StockCode'].isin(top_products.index[:5])]

print("Proposed VIP Bundle based on most popular items:")
for _, row in top_vip_products.iterrows():
    print(f"- {row['Description']} (Code: {row['StockCode']})")
```
