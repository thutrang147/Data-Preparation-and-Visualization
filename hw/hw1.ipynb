##### **Exercise 2: Filtering with Complex Conditions**

```python
# 1
df_reward = df_students[(df_students['math_score'] >= 8) & (df_students['physics_score'] >= 8)]
df_reward
```

```python
# 2
df_needs_improvement = df_students[(df_students['chemistry_score'] < 5) | (df_students['attendance'] == 'Poor')]
df_needs_improvement
```

##### **Exercise 3: Creating New Columns from Arithmetic Operations**

```python
# 1
df_students['average_score'] = round((df_students['math_score'] + df_students['physics_score'] + df_students['chemistry_score']) / 3, 1)
df_students
```

```python
# 2
df_students['math_std_diff'] = df_students['math_score'] - df_students['math_score'].mean()
df_students
```

##### **Exercise 4: Sorting and Filtering**

##### **Exercise 5: Applying Functions with .map()**

```python
# 1
attendance_status = {
    'Good': 10,
    'Fair': 8,
    'Average': 6,
    'Poor': 3
}
```

##### **Exercise 9: Arithmetic with fill_value**

```python
# 1, 2, 3
math_bonus = pd.Series([0.5, 0.25, 0.5, 0, 0.25])
physics_bonus = pd.Series([0.25, 0.5, 0.5, 0.25, 0.5], index=range(5, 10))
```

```python
# 4
df_students['new_math_score'] = df_students['math_score'].add(math_bonus, fill_value=0)
df_students
```

```python
# 5
df_students['new_physics_score'] = df_students['physics_score'].add(physics_bonus, fill_value=0)
df_students
```
