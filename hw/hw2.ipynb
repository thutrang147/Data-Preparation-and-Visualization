```python
# Create DataFrames

df_orders = pd.read_csv(StringIO(order_data))

df_more_orders = pd.read_csv(StringIO(more_order_data))

df_customers = pd.read_csv(StringIO(customer_data))

df_products = pd.read_csv(StringIO(product_data))
```

##### **Exercise 1: Load and Diagnose df_orders**

```python
# 1
df_orders = pd.read_csv(StringIO(order_data), 
                       na_values=['Not Available', 'N/A', '--'])

df_orders
```

```python
# 2
df_orders.info()
```

##### **Exercise 2: Correcting Data Types**

```python
# 1
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'])
df_orders.info()
```

```python
# 2
df_orders['quantity'] = df_orders['quantity'].astype('float64')
df_orders['price_per_item'] = df_orders['price_per_item'].astype('float64')
df_orders.info()
```

##### **Exercise 3: Detecting and Dropping Missing Data**

```python
# 1
df_orders.isnull().sum()
```

```python
# 2
df_orders = df_orders.dropna(subset=['price_per_item'])
df_orders
```

##### **Exercise 4: Filling Missing Data (Imputation)**

```python
# 1
df_orders['quantity'] = df_orders['quantity'].fillna(1)
df_orders
```

```python
# 2
df_orders['quantity'] = df_orders['quantity'].astype('int')
df_orders.info()
```

##### **Exercise 5: Concatenating Datasets**

```python
# 1 2
df_orders = pd.concat([df_orders, df_more_orders], ignore_index=True)
df_orders
```

##### **Exercise 6: Merging Orders with Customers (One-to-Many Join)**

```python
# 1
df_orders = pd.merge(df_orders, df_customers, on='customer_id')
df_orders
```

```python
# 2
df_orders = pd.merge(df_orders, df_customers, on='customer_id', how='left')
df_orders
```

##### **Exercise 7: Merging with Product Details**

```python
# 1 2
df_merged = pd.merge(df_orders, df_products)
df_merged
```

##### **Exercise 8: Finding Unsold Products**

```python
# 1
unsold_products = pd.merge(df_orders[['product_id']],
                           df_products, on='product_id', how='right')
unsold_products = unsold_products[unsold_products['product_name'].isna()]
unsold_products
```

##### **Exercise 9: Dropping Duplicates**

```python
# 1
df_merged = pd.concat([df_merged, df_merged.tail(1)])
df_merged
```

```python
# 2
df_merged = df_merged.drop_duplicates()
df_merged
```

##### **Exercise 10: Final Analysis Questions**

```python
# 1
df_merged['total_price'] = df_merged['quantity'] * df_merged['price_per_item']
df_merged
```

```python
# 2
total_rev_elect = df_merged[df_merged['category'] == 'Electronics']['total_price'].sum()
total_rev_elect
```

##### **Challenge: Try to solve the following question without using .groupby().**

```python
# 3
customer_spending = {}
for customer_id in df_merged['customer_id'].unique():
    customer_spending[customer_id] = df_merged[df_merged['customer_id'] == customer_id]['total_price'].sum()

for customer, spend in customer_spending.items():
    print(f"Customer {customer}: ${spend:,.2f}")
```

```python
# 4
product_rev = {}
for name in df_merged['product_name'].unique():
    product_rev[name] = df_merged[df_merged['product_name'] == name]['total_price'].sum()

max_product = max(product_rev.items(), key=lambda x: x[1])
print(max_product)
```

```python
# 5
category_rev = {}
for name in df_merged['category'].unique():
    category_rev[name] = df_merged[df_merged['category'] == name]['total_price'].sum()

min_category = min(category_rev.items(), key=lambda x: x[1])
print(min_category)
```

```python
# 6
city_rev = {}
for name in df_merged['city'].unique():
    city_rev[name] = df_merged[df_merged['city'] == name]['total_price'].mean()

highest_avg_city = max(city_rev.items(), key=lambda x: x[1])
print(highest_avg_city)
```
